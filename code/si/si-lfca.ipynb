```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import CenteredNorm
import matplotlib.path as mpath
import xarray as xr
from cartopy import crs as ccrs, feature as cfeature
import nc_time_axis
```

```python
def add_cyclic_point(xarray_obj, dim='lon', period=360):
    if period is None:
        period = xarray_obj.sizes[dim] / xarray_obj.coords[dim][:2].diff(dim).item()
    first_point = xarray_obj.isel({dim: slice(1)})
    first_point.coords[dim] = first_point.coords[dim]+period
    return xr.concat([xarray_obj, first_point], dim=dim)
```

```python
hr_so_jlfca = xr.open_dataset('/roselab_rit/rford/iHESP-data/iHESP-HR.PICTRL.SO-JLFCA.1x1.nc')
hr_nso_jlfca = xr.open_dataset('/roselab_rit/rford/iHESP-data/iHESP-HR.PICTRL.NSO-JLFCA.1x1.nc')
```

```python
fig1, ax1 = plt.subplots(3, 5, dpi=200, figsize=(10, 5), layout='constrained')

for i, ax in enumerate(ax1[0]):
    ax.contourf(hr_so_jlfca.lon, hr_so_jlfca.lat, hr_so_jlfca.LFP.isel(mode=i, var=0), levels=np.arange(-1, 1.1, 0.1), cmap='RdBu_r', extend='both')
    ax.set_title('')
    ax.set_facecolor('0.3')


for i, ax in enumerate(ax1[1]):
    plot = ax.contourf(hr_so_jlfca.lon, hr_so_jlfca.lat, hr_so_jlfca.LFP.isel(mode=i, var=1), levels=np.arange(-1, 1.1, 0.1), cmap='RdBu_r', extend='both')
    ax.set_title('')
    ax.set_facecolor('0.3')

for i, ax in enumerate(ax1[2]):
    ax.plot(hr_so_jlfca.time, hr_so_jlfca.LFC.isel(mode=i, var=0), color='k', linewidth=1)
    ax.set_title('')

cbar_ax1 = fig1.add_axes([0.357, 1.05, 0.3, 0.025])
fig1.colorbar(plot, cax=cbar_ax1, orientation='horizontal')

ax1[2, 2].set_xlabel('Year')
ax1[0, 0].set_ylabel('Latitude')
ax1[1, 0].set_ylabel('Latitude')
ax1[1, 2].set_xlabel('Longitude')
ax1[2, 0].set_ylabel('LFC')
ax1[0, 4].set_ylabel('SST', labelpad=-150.0, fontweight='bold', size=12)
ax1[1, 4].set_ylabel('SSS', labelpad=-150.0, fontweight='bold', size=12)
ax1[0, 0].set_title('LFP/LFC1', fontweight='bold', size=12)
ax1[0, 1].set_title('LFP/LFC2', fontweight='bold', size=12)
ax1[0, 2].set_title('LFP/LFC3', fontweight='bold', size=12)
ax1[0, 3].set_title('LFP/LFC4', fontweight='bold', size=12)
ax1[0, 4].set_title('LFP/LFC5', fontweight='bold', size=12)
```
